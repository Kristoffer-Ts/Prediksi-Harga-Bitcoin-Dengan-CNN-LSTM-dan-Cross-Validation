```python
# Folds
folds = np.arange(1, len(CNN_LSTM_2_layer) + 1)

# Plotting
plt.figure(figsize=(12, 6))

# Plot CNN
cnn_folds, cnn_scores = [], []
for i in range(len(CNN_LSTM_2_layer)):
    if CNN_LSTM_2_layer[i] >= 0:
        plt.plot(folds[i], CNN_LSTM_2_layer[i], 'bo')
        plt.annotate(f'{CNN_LSTM_2_layer[i]:.3f}', (folds[i], CNN_LSTM_2_layer[i]), textcoords="offset points", xytext=(0, -25), ha='center', color='blue')
        cnn_folds.append(folds[i])
        cnn_scores.append(CNN_LSTM_2_layer[i])

plt.plot(cnn_folds, cnn_scores, 'b-', label='CNN-LSTM 2 Layers')

# #LSTM 
# lstm_folds, lstm_scores = [], []
# for i in range(len(LSTM)):
#     if LSTM[i] >= 0:
#         plt.plot(folds[i], LSTM[i], 'rs')
#         plt.annotate(f'{LSTM[i]:.3f}', (folds[i], LSTM[i]), textcoords="offset points", xytext=(0,-15), ha='center', color='red')
#         lstm_folds.append(folds[i])
#         lstm_scores.append(LSTM[i])

# plt.plot(lstm_folds, lstm_scores, 'r-', label='LSTM')

# Plot CNN-LSTM
cnn_lstm_folds, cnn_lstm_scores = [], []
for i in range(len(CNN_LSTM_Single_Layer)):
    if CNN_LSTM_Single_Layer[i] >= 0:
        plt.plot(folds[i], CNN_LSTM_Single_Layer[i], 'g^')
        plt.annotate(f'{CNN_LSTM_Single_Layer[i]:.3f}', (folds[i], CNN_LSTM_Single_Layer[i]), textcoords="offset points", xytext=(0, 25), ha='center', color='green')
        cnn_lstm_folds.append(folds[i])
        cnn_lstm_scores.append(CNN_LSTM_Single_Layer[i])

plt.plot(cnn_lstm_folds, cnn_lstm_scores, 'g-', label='CNN-LSTM-Single Layer')

plt.xlabel('Fold')
plt.ylabel('R2_score')
plt.title('Perbandingan CNN-LSTM 1 layer CNN dan CNN-LSTM 2 Layer CNN pada epoch 100')
plt.yscale('linear')  # Use linear scale for R2 score
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
